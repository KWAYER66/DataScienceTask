```python
listverzeichnis = os.listdir('.')
```

```python
#Ersten Eindruck verschaffen
for datei in listverzeichnis:
    if 'ONTIME_REPORTING-001.csv' in datei:
        datei1= pd.read_csv(datei)
        print(datei1.shape)
        spalten= set(datei1.columns)
        print(spalten)
```

```python
#Flüge Airlines
datei1['OP_UNIQUE_CARRIER'].value_counts().plot(kind='bar')
```

```python
#Flüge Abflughafen
#datei1['ORIGIN'].value_counts().plot(kind='bar')
datei1['ORIGIN'].value_counts()[datei1['ORIGIN'].value_counts()>5000].plot(kind='bar')
```

```python
#Flüge Zielflughafen
#datei1['DEST'].value_counts().plot(kind='bar')
datei1['DEST'].value_counts()[datei1['DEST'].value_counts()>5000].plot(kind='bar')
```

```python
#Verspätungen nach Gruppen Abflug, -1: Vorzeitig, Andere Gruppen 15min Zeitslots
datei1['DEP_DELAY_GROUP'].value_counts().plot(kind='bar')
```

```python
#Verspätungen nach Gruppen Ankunft, -1: Vortzeitg, Andere Gruppen 15min Zeitslots
datei1['ARR_DELAY_GROUP'].value_counts().plot(kind='bar')
```

```python
#Verspätungsgründe
datei1['CANCELLATION_CODE'].value_counts().plot(kind='bar')
#A= Carrier, B= Weather, C= NAS, D = Security
```

```python
klist = [(i, j) for i in range(3) for j in range(4)]
```

```python
fig, axes = plt.subplots(figsize =(24,10),nrows=3,ncols=4, sharey=True)
k = 0
for datei in listverzeichnis:
    if 'ONTIME_REPORT' in datei:
        i, j = klist[k]
        df = pd.read_csv(datei)
        df2 = pd.pivot_table(df, index='OP_UNIQUE_CARRIER', values=['TAXI_OUT'], aggfunc=np.mean)
        k = k + 1
        df2.sort_values(by='TAXI_OUT').plot(kind='bar', ax = axes[i, j])
        df2.sort_values(by='TAXI_OUT').plot(kind='line', ax = axes[i, j], color='g')
        
        
```
